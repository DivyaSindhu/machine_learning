insert a logistic regression classifier to predict whether a flower is virginica or not

```python
from sklearn import datasets
iris = datasets.load_iris()
print(list(iris.keys()))
```

```python
iris['data']
```

```python
iris['target']
```

```python
print(iris['DESCR'])
```

three types of flowers based on the length of petals and sepals
Iris-Setosa\n                - Iris-Versicolour\n                - Iris-Virginica\n  

```python
print(iris['data'].shape)
```

x is label and y is target

```python
x = iris["data"][:,3:]
```

```python
print(iris['data'])
```

```python
x
```

```python
y = (iris['target'] == 2)
```

it will return true or false.

```python
y
```

if we want to return 0 0r 1 then use astype

```python
import numpy as np
y = (iris['target'] == 2).astype(np.int)
y
```

Train a logistic regression classifier

```python
from  sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(x,y)
```

```python
ex1 = clf.predict(([[1.6]]))
```

```python
ex1
```

```python
ex2 = clf.predict(([[2.6]]))
```

```python
ex2
```

0 means flower is not virginica and 1 means flower is virginica

now plot the visualization using Matplotlib as below

```python

x_new = np.linspace(0,3,1000).reshape(-1,1)
x_new
```

```python
y_prob = clf.predict_proba(x_new)
y_prob
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x_new, y_prob[:,1], "b-", label="virginica" )
plt.show()
```
